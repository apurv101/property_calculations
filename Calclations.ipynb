```python
import pandas as pd
df = pd.read_csv("reportallusa_2017-12-08_pSljv5ArqzsZkodC1baRgrkpC.csv")
```

```python
#Exctracting data points with muni_name Berkeley
df = df[df.muni_name == "Berkeley"]
#Removing data points with market value 0
df = df[df.mkt_val_tot != 0]
#Removing data points with address null
df = df[df.address.notnull()]
```

```python
#Taking out only these four fields
df_rel = df[["address", "muni_name", "land_use_code", "mkt_val_tot"]]
```

```python
#All unique land use codes
df["land_use_code"].unique()
```

```python
# Land codes starting with "1" are residential
df_rel_code = df_rel[df_rel.land_use_code.astype('str').str[0] == "1"]
```

```python
# Only residential data of Berkeley
df_rel_code
```

```python
adu_size = 450
multipliers = {"detached_ADU":250,"attached_ADU":200, "interior_conversion":125, "basement_conversion":145,
               "garage_conversion":135}
city_multiplier = 1.656785149
rental_value = 3.35
market_appreciation = 1.05
```

```python
multipliers.keys()
```

```python
df_rel_code["rental_income"] = rental_value * adu_size
```

```python
#Find appreciation value for all kind of ADUs and add them to dataframe
for i in multipliers.keys():
    df_rel_code["AV_"+ i] = adu_size * multipliers[i] * city_multiplier
    df_rel_code["Appreciation_%_for_"+ i] = adu_size * multipliers[i] * city_multiplier / df_rel_code.mkt_val_tot
    df_rel_code["5_year_ROI_for_" + i] = (adu_size * multipliers[i] * city_multiplier * market_appreciation**5) + (df_rel_code.rental_income*12*5)
    df_rel_code["mortgage_payment_for_"+ i] = adu_size * multipliers[i] * city_multiplier / 260
```

```python
df_rel_code.to_csv("calculated_data_berkeley.csv")
```

# Importing all the data

```python
import pandas as pd
df = pd.read_csv("reportallusa_2017-12-08_pSljv5ArqzsZkodC1baRgrkpC.csv")
```

```python
df.dtypes
```

## Extracting the data only for Berkeley

```python
#Exctracting data points with muni_name Berkeley
df = df[df.muni_name == "Berkeley"]
```

## Extracting only the relevant columns 

```python
df_relevant = df[["address", "land_use_code", "bldg_sqft", "mkt_val_tot", "calc_acreage", "lat", "lon"]]
```

```python
df_relevant
```

## Total data points are 19972

```python
df_relevant.shape
```

## Removing data points
### Market value is 0
### Address is null
### Land_use_code is not 110

```python
#Removing data points with market value 0
df_relevant = df_relevant[df_relevant.mkt_val_tot != 0]
#Removing data points with address null
df_relevant = df_relevant[df_relevant.address.notnull()]

#Extracting data with land_use_code "110"...only residential data
df_relevant = df_relevant[df_relevant.land_use_code.astype('str').str[0:3] == "110"]
```

### Adding a column "lot_size_sqft" which is 43560 times of calc_acreage

```python
#Creating a new column of lot size
df_relevant["lot_size_sqft"] = df_relevant.calc_acreage * 43560
```

## Adding a new column "far" which is "floor / lot"

```python
# Creating a new column with floor area ratio = bldg_sqft / lot_size_sqft
df_relevant["far"] = df_relevant.bldg_sqft / df_relevant.lot_size_sqft
```

```python
df_relevant
```

## Extracting the properties where far >= 1

```python
# Property with far greater than 1
df_relevant_far_gt1 = df_relevant[df_relevant.far >= 1]
```

```python
# We can only do a conversion here
df_relevant_far_gt1
```

## Extracting the properties where far < 1

```python
# Property with far less than 1
df_relevant_far_lt1 = df_relevant[df_relevant.far < 1]
```

```python
# we can add extra square footage (attached or detached ADU)
df_relevant_far_lt1
```

## How many extra square foot we can add?

```python
# Rule 1: adu + main residence cannot be bigger than lot size
# adu_size_rule_1 = lot - floor

# Rule 2: adu cannot be bigger than 75% of main residence
# adu_size_rule_2 = floor * 3 / 4

# Rule 3: adu cannot be bigger than 750 sf
# adu_size_rule_3 = 750

df_relevant_adu_size_rule = df_relevant.copy()
```

```python
df_relevant_adu_size_rule["adu_size_rule_1"] = df_relevant_adu_size_rule.lot_size_sqft - df_relevant_adu_size_rule.bldg_sqft	
```

```python
df_relevant_adu_size_rule["adu_size_rule_2"] = df_relevant_adu_size_rule.bldg_sqft * 3 / 4
```

```python
df_relevant_adu_size_rule["adu_size_rule_3"] = 750
```

```python
#Calculating maximum adu size allowed(smallest of three rules)
df_relevant_adu_size_rule["max_adu_size_allowed"] = df_relevant_adu_size_rule.loc[:, ['adu_size_rule_1', 'adu_size_rule_2', 'adu_size_rule_3']].min(axis=1)
```

```python
df_relevant_adu_size_rule
```

# Narrowing it down even more

```python
df_relevant_narrow = df_relevant_adu_size_rule.copy()
```

## Extracting properties which have far < 0.4

```python
# Properties if Far is less than 0.4
df_relevant_narrow =df_relevant_narrow[df_relevant_narrow.far < 0.4]
```
